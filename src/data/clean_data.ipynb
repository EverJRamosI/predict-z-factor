```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Factor_Z.csv')
df.head(6)
```

```python
df_c = df.copy()
df_c.head()
```

```python
df_c.index.name = 'Pressure(psia)'
df_c.head()
```

```python
df_c.set_index('Unnamed: 0', inplace=True)
df_c.head()
```

```python
df_c.index.name = 'Pressure(psia)'
df_c.head()
```

```python
plt.figure(figsize=(12, 8))

for col in df_c.columns:
    label = f'{col}'
    plt.plot(df_c.index, df_c[col], label=label)
    
plt.xlabel('Pressure(psia)')
plt.ylabel('Factor Z')
plt.legend(fontsize=8, loc='best')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df_cn = df_c[~(df_c < 0).any(axis=1)]
df_cn.head()
```

```python
plt.figure(figsize=(12, 8))

for col in df_cn.columns:
    label = f'{col}'
    plt.plot(df_cn.index, df_cn[col], label=label)
    
plt.xlabel('Pressure(psia)')
plt.ylabel('Factor Z')
plt.legend(fontsize=8, loc='best')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df_cn.to_csv('Factor_Zc.csv')
```
